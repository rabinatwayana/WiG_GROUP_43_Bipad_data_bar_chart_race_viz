```python
import bar_chart_race as bcr
import pandas as pd
import warnings

# Suppress the specific UserWarning from bar_chart_race
warnings.filterwarnings("ignore", category=UserWarning, module='bar_chart_race')

# Load your DataFrame
# df = pd.read_csv("/Users/rabina/Projects/Bipad_data_extraction/land-cover-merged.csv", index_col="date", parse_dates=True)

sheets_dict = pd.read_excel("/Users/rabina/Projects/Bipad_data_extraction/province_wise_hazard_combined_sheet.xlsx", sheet_name=None)


province_name={
    "province_1":"Trend of Hazards (Koshi Province)",
    "province_2":"Trend of Hazards (Madhesh Province)",
    "province_3":"Trend of Hazards (Bagmati Province)",
    "province_4":"Trend of Hazards (Gandaki Province)",
    "province_5":"Trend of Hazards (Lumbini Province)",
    "province_6":"Trend of Hazards (Karnali Province)",
    "province_7":"Trend of Hazards (Sudurpashchim Province)",
 }

# Loop over each sheet and perform actions
for sheet_name, df in sheets_dict.items():

    # Set the 'date' column as the index and parse dates
    df.set_index('year', inplace=True)
    df.index = pd.to_datetime(df.index, format='%Y')

    # Create the bar chart race video
    bcr.bar_chart_race(
        df=df,
        filename=f'{sheet_name}.mp4',
        orientation='h',
        sort='desc',
        n_bars=5,
        fixed_order=False,
        fixed_max=True,
        steps_per_period=10,
        interpolate_period=True,
        label_bars=True,
        bar_size=.75,
        period_label={'x': .99, 'y': .28, 'ha': 'right', 'va': 'center','size':7},
        # period_fmt='%Y-%m-%d, ',
        period_fmt='%Y',
        period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                        's': f'Total count: {v.nlargest(6).sum():,.0f}',
                                        'ha': 'right', 'size': 8, 'family': 'Courier New','size':6},
        # perpendicular_bar_func='median',
        period_length=1500,
        figsize=(3, 2.25),
        dpi=500,  # Adjusted DPI
        cmap='dark12',  #accent
        title=f'{province_name[sheet_name]}',
        title_size='7',
        bar_label_size=6,
        tick_label_size=6,
        scale='linear',
        writer=None,
        bar_kwargs={'alpha': .7},
        filter_column_colors=False,
        # xlabel='Year',    # Add x-axis label
        # ylabel='Hazard'  
    )
```
